```python
df = pd.read_csv(files[0],usecols=['Unnamed: 0','reg','topic','eng','dial','bleu','pred'], index_col = 'Unnamed: 0')
res = re.search(r'(?<=_results_)(.*)(?=\_final.*)',files[0]).group()
df.rename(columns={'bleu':'bleu_msp8000','pred' : 'pred_msp8000'}, inplace=True)

for file in files[1:] :
    res = re.search(r'(?<=_results_)(.*)(?=\_final.*)',file).group()
    temp_ = pd.read_csv(file, usecols=['Unnamed: 0','reg','topic','eng','dial','bleu','pred'], index_col = 'Unnamed: 0')
    temp_.rename(columns={'bleu':f'bleu_{res}','pred' : f'pred_{res}'},inplace=True)
    df = pd.merge(df, temp_.iloc[:,[-1,-2]], right_index=True, left_index=True)
# reorder_col = ['reg','topic','eng','dial','pred_spm','pred_msp','pred_cmsp','bleu_spm','bleu_msp','bleu_cmsp']
# df = df[reorder_col].copy()
```

$$
r = (I_n - (.85)M^T)^{-1}\biggl(\frac{1 - (.85)}{5}\biggr)\cdot 1 
$$

$$
x=1 , if\  a <= 5 ; x=2 , if\  5 < a <= 10 ;
x=3 , if\  10 < a <= 15 ; x=4 , if\  15 < a <= 20;\\
x=5 , if\  20 < a <= 25; x=6 , if\  a > 25\\
$$


```python
df['tok_len'] = df['eng'].apply(lambda x : len(x.split()))
conditions = [
    (df['tok_len'] <= 5),
    (df['tok_len'] > 5) & (df['tok_len'] <= 10),
    (df['tok_len'] > 10) & (df['tok_len'] <= 15),
    (df['tok_len'] > 15) & (df['tok_len'] <= 20),
    (df['tok_len'] > 20) & (df['tok_len'] <= 25),
    (df['tok_len'] > 25)
    ]
values = [1, 2, 3, 4,5,6]
```

```python
df['tok_cat'] = np.select(conditions, values)
```

```python
#8000
df_ev = pd.read_excel(files1[0], index_col = 'Unnamed: 0')

for i,file in enumerate(files1[1:]) :
    i +=1
    use_cols=['Unnamed: 0','sat_eval1', 'acc_eval1','sat_eval2','acc_eval2','sat_eval3', 'acc_eval4']
    df_ev_temp = pd.read_excel(file, usecols = use_cols, index_col = 'Unnamed: 0')
    df_ev_temp.rename(columns = {'sat_eval1': f'sat_eval1_{i}', 'acc_eval1':f'acc_eval1_{i}', 'sat_eval2': f'sat_eval2_{i}', 'acc_eval2':f'acc_eval2_{i}','sat_eval3': f'sat_eval3_{i}', 'acc_eval4':f'acc_eval3_{i}'},inplace=True)
    df_ev = pd.merge(df_ev, df_ev_temp, right_index =True, left_index=True)
    
new_col_order =['eng', 
'pred_spm', 'sat_eval1', 'sat_eval1_1','sat_eval1_2','sat_eval1_3','sat_eval1_4', 
            'acc_eval1', 'acc_eval1_1','acc_eval1_2','acc_eval1_3','acc_eval1_4',
'pred_msp', 'sat_eval2', 'sat_eval2_1','sat_eval2_2','sat_eval2_3','sat_eval2_4', 
            'acc_eval2', 'acc_eval2_1','acc_eval2_2','acc_eval2_3','acc_eval2_4',
'pred_cmsp','sat_eval3', 'sat_eval3_1','sat_eval3_2','sat_eval3_3','sat_eval3_4',  
            'acc_eval4', 'acc_eval3_1','acc_eval3_2','acc_eval3_3','acc_eval3_4']

df_ev['spm_sat_mean_8000'] = df_ev[['sat_eval1', 'sat_eval1_1','sat_eval1_2','sat_eval1_3','sat_eval1_4']].mean(axis=1)
df_ev['spm_acc_mean_8000'] = df_ev[['acc_eval1', 'acc_eval1_1','acc_eval1_2','acc_eval1_3','acc_eval1_4']].mean(axis=1)
df_ev['msp_sat_mean_8000'] = df_ev[['sat_eval2', 'sat_eval2_1','sat_eval2_2','sat_eval2_3','sat_eval2_4']].mean(axis=1)
df_ev['msp_acc_mean_8000'] = df_ev[['acc_eval2', 'acc_eval2_1','acc_eval2_2','acc_eval2_3','acc_eval2_4']].mean(axis=1)
df_ev['cmsp_sat_mean_8000'] = df_ev[['sat_eval3', 'sat_eval3_1','sat_eval3_2','sat_eval3_3','sat_eval3_4']].mean(axis=1)
df_ev['cmsp_acc_mean_8000'] = df_ev[['acc_eval4', 'acc_eval3_1','acc_eval3_2','acc_eval3_3','acc_eval3_4']].mean(axis=1)

final_reorder = ['pred_spm',
'spm_sat_mean_8000','spm_acc_mean_8000',
'pred_msp',
'msp_sat_mean_8000','msp_acc_mean_8000',
'pred_cmsp',
'cmsp_sat_mean_8000','cmsp_acc_mean_8000',]

final_reorder_only_values = [
'spm_sat_mean_8000','spm_acc_mean_8000',
'msp_sat_mean_8000','msp_acc_mean_8000',
'cmsp_sat_mean_8000','cmsp_acc_mean_8000',]

final_df = pd.merge(df, df_ev[final_reorder_only_values], right_index=True, left_index=True)

tokenizers = ['spm','msp','cmsp']
vocab_size = '8000'
eval_1 = final_df.loc[:,[f'bleu_{tokenizers[0] + vocab_size}',f'{tokenizers[0]}_sat_mean_{vocab_size}', f'{tokenizers[0]}_acc_mean_{vocab_size}']].agg('mean').to_frame().transpose().rename(index={0:tokenizers[0]+'8k'},columns={f'bleu_{tokenizers[0] + vocab_size}':'bleu',f'{tokenizers[0]}_sat_mean_{vocab_size}':'sat_qual_mean',f'{tokenizers[0]}_acc_mean_{vocab_size}':'acc_qual_mean'})
for tokenizer in tokenizers[1:] :
    temp = final_df.loc[:,[f'bleu_{tokenizer + vocab_size}',f'{tokenizer}_sat_mean_{vocab_size}', f'{tokenizer}_acc_mean_{vocab_size}']].agg('mean').to_frame().transpose().rename(index={0:tokenizer+'8k'},columns={f'bleu_{tokenizer + vocab_size}':'bleu',f'{tokenizer}_sat_mean_{vocab_size}':'sat_qual_mean',f'{tokenizer}_acc_mean_{vocab_size}':'acc_qual_mean'})
    eval_1 = pd.concat([eval_1,temp],axis=0)
```
