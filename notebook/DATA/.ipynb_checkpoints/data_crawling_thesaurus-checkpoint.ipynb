## Crawling_preprocess

```python
def preprocess(local_url, local_name):
    df_text = []
    url = local_url
    page = requests.get(url)
    soup = bs(page.text, "html.parser")
    sentence = soup.section.ul.get_text()
    df_text.append(sentence)
    
    for s in range(2, 5000):
        test_url = f'{local_url}?%EC%AA%BD={s}'
        page_test = requests.get(test_url)
        soup_test = bs(page_test.text, "html.parser")
        sentence_test = soup_test.section.ul.get_text()
    
        if sentence[:10] == sentence_test[:10]:
            print("모두 완료 되었습니다. 작업을 종료합니다.")
            break
        else:
            df_text.append(sentence_test)
            print(f'작업 완료 : {s} 페이지까지' )
    
    final = ''.join(df_text)
    sentence = final.split('\n\n\n\n ')
#     return sentence
    text = []
    for i in sentence :
        i = re.sub('[\n‘’]','',i)
        i = i.strip()
        x,y = i.split(':')
        x = re.sub('\(.*\)','',x)
        text.append((x,y))
    
    df = pd.DataFrame(text,columns =['dial','meaning'])
    df['original'] = df['meaning'].apply(lambda x : x.split('의 방언')[0] if '의 방언' in x else np.NaN)
    df['reg'] = f'{local_name}'
    df = df[['dial','original','meaning','reg']]
    return df
```

```python
test = ['나는 나의 의 방언이야', '너는 너의 의 방언이야.', '의 방언']
for i in test:
    if '의 방언' in i:
        print(i)
```
